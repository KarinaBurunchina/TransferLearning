```python
import numpy as np 
import pandas as pd
```

```python
data2 = pd.read_csv('child_combined.csv', low_memory = False)
```

```python
#delete rows with repeated headers
rows_del = data2.index[data2['timestamp'] == 'timestamp'].tolist()
data2 = data2.drop(rows_del)
```

```python
#replace Nan by 0
data2 = data2.fillna(0)
```

```python
#drop rows with full of 0. Where frame is '-1' 
data_without0 = data2[~data2.purple_frame_idx.str.contains("-1", na=False)]

#remove all columns between column index 192 to 442
new_data = data_without0.drop(data_without0.columns[192:443], axis = 1)
new_data.info()
```

```python
#drop first 10 columns and purple_child_task_engagement column.
new_data = new_data.drop(new_data.columns[0:10], axis = 1)
new_data = new_data.drop("purple_child_task_engagement", axis = 1)
```

```python
new_data.columns.tolist()
```

```python
updated_data = new_data.drop(new_data.columns[183:187], axis = 1)
```

```python
updated_data.columns[176]
```

```python
updated_data = updated_data.drop(updated_data.columns[176:182], axis = 1)
len(updated_data.columns.tolist())
```

```python
updated_data.columns.tolist()
```

```python
#values in pinsoro are strings, we replace by floats

updated_data['purple_child_social_engagement'] = updated_data['purple_child_social_engagement'].replace(regex='solitary', value=1)
updated_data['purple_child_social_engagement'] = updated_data['purple_child_social_engagement'].replace(regex='onlooker', value=2)
updated_data['purple_child_social_engagement'] = updated_data['purple_child_social_engagement'].replace(regex='parallel', value=3)
updated_data['purple_child_social_engagement'] = updated_data['purple_child_social_engagement'].replace(regex='associative', value=4)
updated_data['purple_child_social_engagement'] = updated_data['purple_child_social_engagement'].replace(regex='cooperative', value=5)
```

```python
for col in range(0, 177):
    updated_data[updated_data.columns[col]] = pd.to_numeric(updated_data[updated_data.columns[col]])
```

```python
updated_data.info()
```

```python
updated_data.to_csv("combine_child_upgraded.csv", index = False, header=True)
```

```python
data = pd.read_csv('qamqor.csv')
```

```python
data.sort_values(by = 'childID', ascending=True, inplace=True)
```

```python
data.columns.tolist()
```

```python
#Dropping columns with OpenPose c-values
data_without_c = data.drop(columns=data.columns[5:414:3])
```

```python
#Dropping first 3 columns
data_without_c = data_without_c.drop(data_without_c.columns[0:3], axis = 1)
```

```python
#Dropping columns with OpenPose key points of legs: 19-24

data_without_leg = data_without_c
data_without_leg = data_without_leg.drop(data_without_leg.columns[38:50], axis = 1)
data_without_leg.columns.tolist()
```

```python
#Dropping column with OpenPose key point 8.
data_qamqor_17 = data_without_leg.drop(data_without_leg.columns[16:18], axis = 1)
data_qamqor_17.columns.tolist()
```

```python
#Dropping columns with OpenPose hand key points
data_qamqor_17 = data_qamqor_17.drop(data_qamqor_17.columns[36:120], axis = 1)
data_qamqor_17.columns.tolist()
```

```python
#Dropping last 2 columns
data_qamqor_17 = data_qamqor_17.drop(data_qamqor_17.columns[176:178], axis = 1)
data_qamqor_17.columns.tolist()
```

```python
#re-arranging the columns as follows: 
#face key points -> skeleton -> engagement level

arr = list(range(36, 176))+list( range (0,36))
arr.append(176)
data_qam = data_qamqor_17.iloc[:,arr]
```
